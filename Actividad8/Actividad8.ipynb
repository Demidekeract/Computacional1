```python
def vectorfield(w, t, p):
    """
    Defines the differential equation for the Van der Pol oscillation system.

    Arguments:
        w :  vector of the state variables:
                  w = [x,y]
        t :  time
        p :  vector of the parameters:
                  p = [mu]
    """
    x, y = w

    # Create f = (x',y'):
    f = [y,
         p*(1.0 - x**2.0) * y - x]
    return f
```

```python
# Use ODEINT to solve the differential equations defined by the vector field
from scipy.integrate import odeint
import numpy as np

# Parameter values
# Damping coefficient:
mu = 0.5
# Initial conditions
# x is the initial displacement and y is the initial velocity
x = 2.0
y = -3.0

stoptime = 50.0
numpoints = 2500

# Create the time samples for the output of the ODE solver.
# I use a large number of points, only because I want to make
# a plot of the solution that looks nice.
t = [stoptime * float(i) / (numpoints - 1) for i in range(numpoints)]

# Pack up the parameters and initial conditions:
w0 = [x, y]

# Call the ODE solver.
wsol = odeint(vectorfield, w0, t, args=(mu,))

with open('act8-1.dat', 'w') as f:
    # Print & save the solution.
    for t1, w1 in zip(t, wsol):
        print (t1, w1[0], w1[1], file=f)
```

```python
import numpy as np
def forced(w, t, p):
    """
    Defines the differential equation for the Van der Pol oscillation system.

    Arguments:
        w :  vector of the state variables:
                  w = [x,y]
        t :  time
        p :  vector of the parameters:
                  p = [mu]
    """
    x, y = w
    
    A, om, mu = p

    # Create f = (x',y'):
    f = [y,
         mu*(1.0 - x**2.0) * y - x + A*np.sin(om*t)]
    return f
```

```python
# Use ODEINT to solve the differential equations defined by the vector field
from scipy.integrate import odeint
from numpy import *

# Parameter values
# Damping coefficient:
mu = 8.53
A = 1.2
om = 2.0*pi/10.0
# Initial conditions
# x is the initial displacement and y is the initial velocity
x = 1.0
y = -10.0

stoptime = 1000.0
numpoints = 2500

# Create the time samples for the output of the ODE solver.
# I use a large number of points, only because I want to make
# a plot of the solution that looks nice.
t = [stoptime * float(i) / (numpoints - 1) for i in range(numpoints)]

# Pack up the parameters and initial conditions:
w0 = [x, y]

p=[A,om,mu]

# Call the ODE solver.
wsol = odeint(forced, w0, t, args=(p,))

with open('act8-chaos.dat', 'w') as f:
    # Print & save the solution.
    for t1, w1 in zip(t, wsol):
        print (t1, w1[0], w1[1], file=f)
```
