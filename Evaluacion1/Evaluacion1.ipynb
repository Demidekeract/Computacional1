```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_1 = pd.read_csv("dfsargento.csv", header=None, names=['#', 'Date', 'Pres', 'Temp', 'WaterLvl'])
```

```python
df_1.head()
```

```python
df_2 = pd.read_csv("dfsargento-salinidad.csv", header=None, names=['#','Date','Cond', 'Temp','SpecCond','Sal'])
```

```python
df_2.head()
```

```python
df_1['Ndate'] = pd.to_datetime(df_1['Date'], format='%m/%d/%Y %H:%M:%S')
df_1['Month'] = df_1['Ndate'].dt.month
df_1.head()
```

```python
df_2['Ndate'] = pd.to_datetime(df_2['Date'], format='%m/%d/%Y %H:%M:%S')
df_2['Month'] = df_2['Ndate'].dt.month
df_2.head()
```

```python
ax = sns.boxplot(x="Month", y="WaterLvl", data=df_1)
plt.show()
df_1.describe()
```

```python
ax = sns.boxplot(x="Month", y="Sal", data=df_2)
plt.show()
df_2.describe()
```

```python
ax = sns.boxplot(x="Month", y="Temp", data=df_1)
plt.show()
df_1.describe()
```

```python
ax = sns.boxplot(x="Month", y="Temp", data=df_2)
plt.show()
df_2.describe()
```

```python
dfa=pd.DataFrame(df_1)
```

```python
dfb=pd.DataFrame(df_2)
```

```python
df_3=pd.concat([dfa.WaterLvl,dfb.Sal], axis=1)
```

```python
df_3.head()
```

```python
sns.set(style="darkgrid", color_codes=True)

g = sns.jointplot("WaterLvl", "Sal", data=df_3, kind="reg",
                   color="r", size=7)
plt.show(g)
```

```python
sns.set(style="darkgrid", color_codes=True)

g = sns.jointplot("WaterLvl", "Temp", data=df_1, kind="reg",
                   color="r", size=7)
plt.show(g)
```

```python
sns.set(style="darkgrid", color_codes=True)

g = sns.jointplot("Sal", "Temp", data=df_2, kind="reg",
                   color="r", size=7)
plt.show(g)
```

```python
plt.figure(); df_1.WaterLvl.plot(); plt.legend(loc='best')
plt.title("Variación del nivel del mar")
plt.ylabel("Nivel (Metros)")
plt.grid(True)
plt.show()
```

```python
plt.figure(); df_2.Sal.plot(); plt.legend(loc='best')
plt.title("Variación de la salinidad")
plt.ylabel("Concentración (ppt)")
plt.grid(True)
plt.show()
```

```python
plt.figure(); df_1.Temp.plot(); plt.legend(loc='best')
plt.title("Variación del nivel del mar")
plt.ylabel("Nivel (Metros)")
plt.grid(True)
plt.show()
```
