```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
something=lambda x: x in [0, 1, 2, 4, 5]
AlbDec = pd.read_csv('Albuquerque_dec_22.txt', skiprows=something, sep='\s+')
AlbJun = pd.read_csv('Albuquerque_jun_22.txt', skiprows=something, sep='\s+')
```

```python
ADec = pd.DataFrame(AlbDec)
```

```python
AJun = pd.DataFrame(AlbJun)
```

```python
ADec1 = ADec[['HGHT','PRES']]
plt.figure(); ADec1.plot(x="PRES"); plt.legend(loc='best') 
plt.axis([0,1000,1500,30000])
plt.title("Variación de la presión")
plt.xlabel("Presión (hPa)")
plt.ylabel("Altura (m)")
plt.grid(True)
plt.show()
```

```python
AJun1 = AJun[['HGHT','PRES']]
plt.figure(); AJun1.plot(x="PRES"); plt.legend(loc='best') 
plt.axis([0,1000,1500,30000])
plt.title("Variación de la presión")
plt.xlabel("Presión (hPa)")
plt.ylabel("Altura (m)")
plt.grid(True)
plt.show()
```

```python
ADec1 = ADec[['HGHT','TEMP']]
plt.figure(); ADec1.plot(x="TEMP"); plt.legend(loc='best') 
plt.axis([-70,0,1500,30000])
plt.title("Variación de la presión")
plt.xlabel("Temperatura (°C)")
plt.ylabel("Altura (m)")
plt.grid(True)
plt.show()
```

```python
AJun1 = AJun[['HGHT','TEMP']]
plt.figure(); AJun1.plot(x="TEMP"); plt.legend(loc='best') 
plt.axis([-80,0,1500,30000])
plt.title("Variación de la presión")
plt.xlabel("Temperatura (°C)")
plt.ylabel("Altura (m)")
plt.grid(True)
plt.show()
```
